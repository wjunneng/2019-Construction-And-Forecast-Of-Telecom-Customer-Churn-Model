```python
%matplotlib inline

def draw(df, fd, **params):
    """
    绘制箱形图
    :param df: 
    :param params: 
    :return: 
    """
    import matplotlib.pyplot as plt 

    for column in df.columns:    
        tmp = pd.concat([df[column], fd[column]], axis=1)
        tmp.boxplot()#这里，pandas自己有处理的过程，很方便哦。
        plt.ylabel("y")
        plt.xlabel("x")#我们设置横纵坐标的标题。
        plt.show()

draw(train_15p, test)
```
